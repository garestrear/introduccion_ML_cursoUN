# <center> 4. Varias líneas en una gráfico
________________
## Para graficar varias líneas en una misma gráfica usamos la función `plt.plot()` tantas veces como líneas queramos graficar.

## <u> Tamaño, Nombre de los ejes, título y guardar la gráfica </u>

## 1. Usamos ` plt.figure(figsize=(10,3)` para una figura de tamaño `(10,3)`.

##2. Usamos `plt.xlabel() ` y  `plt.ylabel()` para poner nombre a los ejes $x$ y $y$, respectivamente.  

## 3. Para  poner **título** iniciamos el código creando un objeto de la clase` plt.figure()`. Luego, podemos añadir un título usando `fig.suptitle()`.

## 4.  Para **guardar** una figura usamos `plt.savefig()`. 


```
# Creamos un objeto de la figura
fig  = plt.figure(figsize =(10,3)) # el parámetro figsize cambia el tamaño de la figura 

# Las siguiente dos líneas de código  generan las líneas en la gráfica 
# La etiqueta se usa cuando se crea una leyenda
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'm*--', label = 'Gráfica 1') 
plt.plot([-1,1,3], [2,6,1.5], 'go-', label = 'Gráfica 2')

##**************************************************************************##
##***********************Información adicional******************************##
##**************************************************************************##
# Etiquetas de los ejes 
plt.ylabel('Nombre eje y') 
plt.xlabel('Nombre eje x')
# Título de la figura
fig.suptitle('Título de la Figura', fontsize =20, color='c')
# Crear una leyenda de la gráfica
# Usar plt.legend(loc= (num, num)) para cambiar la posición de la leyenda
plt.legend() 
# Para guardar la gráfica ejecuta
plt.savefig('Mi primera gráfica en python.png')
##**************************************************************************##
##**************************************************************************##

plt.show()
```

## Esta función devuelve una figura `fig` y un arreglo de subgráficas `axs`.

```
fig, axs = plt.subplots(1,2,figsize=(15,5), sharey=True, sharex=True)
# Primera gráfica
axs[0].plot([1, 2, 3, 4], [1, 4, 9, 16], 'm*--')
# Segunda gráfica
axs[1].plot([-1,1,3], [2,6,1.5], 'go-')

##**************************************************************************##
##***********************Información adicional******************************##
##**************************************************************************##
# Etiquetas de los ejes 
axs[0].set_ylabel('Nombre eje y de la primera gráfica' )
axs[0].set_xlabel('Nombre eje x  de la primera gráfica')
axs[0].set_title('Título de la primera gráfica', c= np.random.rand(3), fontsize = 16)

# Nombre de los ejes y título 
axs[1].set_ylabel('Nombre eje y de la segunda gráfica')
axs[1].set_xlabel('Nombre eje x  de la segunda gráfica')
axs[1].set_title('Título de la segunda gráfica', c= np.random.rand(3), fontsize = 16)

# Título de toda la figura
fig.suptitle('Acá el título de la figura', fontsize = 20, color = 'navy') 
##**************************************************************************##
##**************************************************************************##

plt.show()
```

# <center> 6. Graficando funciones
________________

## Para graficar una función `mi_fun` usamos el comando `plt.plot(x,y)`, donde:`

## 1.  `x = [x_1, x_2, ..., x_d]`  es un arreglo o lista que contiene primera coordenada de puntos a gráficar
## 2.   `y = [mi_fun(x_1), mi_fun(x_2),..., mi_fun(x_d)]` es el arreglo o lista de la imagenes de `mi_fun`.

## Para generar el arreglo `x` usamos la función de numpy `np.linspace()`.

## El comando `np.linspace(a, b, k)` genera $k$ números igualmente espaciados desde $a$ hasta $b$ (incluyéndolos).


```
# Generamos 61 puntos equidistantes entre -3 y 3
x = np.linspace(-3, 3, 61)
print(x)
```

## Grafiquemos la función $x^{4} - 7x^2 + x +2$. 

```
# Definimos la función a graficar
def mi_fun(x):
  return x**4 - 7*x**2 + x + 2
```

## Para definir $y$  usamos `list(map(mi_fun, x))`

```
y = list(map(mi_fun, x))
print(y)
```

```
# Graficamos la función mi_fun desde -3 hasta 3
plt.plot(x, y)
plt.show()
```

## También podemos usar las funciones propias de `numpy`, por ejemplo `np.sin()` genera la función seno.

```
y = np.sin(x)
#Graficamos la función mi_fun desde -3 hasta 3
plt.plot(x, y)
plt.show()
```

## Para fijar el número de barras, por ejemplo a 40, agregamos el argumento `bins= 40.`

## Una fórmula bastante usada para calcular los `bins` es $\sqrt{n}$, siendo $n$ la cantidad de datos que tenemos.

```
temp2 = temp.iloc[10,3:]
n = int(np.sqrt(len(temp2)))

plt.hist(temp2, bins=n, edgecolor='white', density=True) 
# Color y opacidad también pueden ser modificados de manera usual.
# Cambiar el tipo de histograma (opcional):  histtype : {'bar', 'barstacked', 'step', 'stepfilled'}
# Cambiar la orientación (opcional): orientation : {'horizontal', 'vertical'}

# mostramos el histograma
plt.show()
```

```
n, bins, patches = plt.hist(temp2, bins=10, edgecolor='b',
                            linewidth=0.5, alpha=0.7)
                          
n = n.astype('int') # Debe ser un número entero
# Good old loop. Choose colormap of your taste
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
# Make one bin stand out   
patches[0].set_fc('red') # Set color
patches[0].set_alpha(1) # Set opacity

# Agregamos un título a nuestro histograma
plt.title('Temperaturas')

# Agregamos los nombres de los ejes
plt.xlabel('Temperatura') 
plt.ylabel('Cantidad de veces de registros') 

# Mostramos el histograma
plt.show()
```

# <center> 8. Texto en un gráfica </center>
________________

## Para incluir texto en una gráfica `ax` podemos usar la función `ax.text()` o `ax.annotate()`. 

## Documentación de `ax.text()` :https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.text.html

## Documentación de `ax.annotate()`: https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.annotate.html
 

```
# Creamos un arreglo de una sola gráfica 
fig, axs = plt.subplots(1, 1, figsize=(10, 5))

# Gráfico de barras
axs.bar([-1, 0, 1, 2], [2, 1, -1, 1.5], alpha=0.4, color=np.random.rand(3))

# Podemos decidir dónde el eje y tiene etiquetas
#Usamos  set_yticks
axs.set_yticks([-1, 0, 1, 1.5, 2 ])    
#Podemos poner cualquier etiqueta que queramos
axs.set_yticklabels(['-10', '0', '10', '15' '20']) 

# Similarmente, podemos poner etiquetas al eje x
axs.set_xticklabels(['et_1','et_2', 'et_3', 'et_4']) 

# Para colocar texto en la figura 
axs.annotate('¡Una barra negativa!', (1, 0), xytext=(0.8, 0.9),
             textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.15),
             fontsize=16, horizontalalignment='right',
             verticalalignment='top')
axs.set_title('Gráfica de Barras', fontsize =20)

plt.show()
```

## Los histogramas se utilizan para mostrar distribuciones de variables, mientras que los gráficos de barras se utilizan para comparar variables. 

## Los histogramas trazan datos cuantitativos con rangos de datos agrupados en contenedores o intervalos, mientras que los gráficos de barras trazan datos categóricos.
